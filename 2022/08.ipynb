```python
import torch
import numpy as np
import re
import math

file = open("08.txt", "r")
lines = file.readlines()
file.close()

n = len(lines[0].strip())
m = len(list(filter(lambda x: len(x.strip()) > 1, lines)))

mt = torch.zeros([n, m], dtype=torch.int32)
for i in range(n):
  for j in range(m):
    mt[i][j] = int(lines[i][j])
print(mt)

p1 = 0
for i in range(n):
  for j in range(m):
    if i == 0 or j == 0 or i == n - 1 or j == m - 1:
      p1 += 1
    else:
      v1 = mt[i:i+1, 0:j]
      v2 = mt[i:i+1, j+1:]
      v3 = mt[0:i, j:j+1]
      v4 = mt[i+1:, j:j+1]
      cur = mt[i, j]
      if v1.max() < cur or v2.max() < cur or v3.max() < cur or v4.max() < cur:
        p1 += 1
p1

```

```python
dirs = [torch.tensor([-1, 0]), torch.tensor([1, 0]), torch.tensor([0, -1]), torch.tensor([0, 1])];
dims = torch.tensor([n, m])

p2 = 0
for i in range(n):
  for j in range(m):
    coord = torch.tensor([i, j])
    h = mt[i, j]
    cellScore = 1
    for d in dirs:
      dScore = 0
      for step in range(1, n+m):
        cur = coord + d * step
        if torch.any(cur < 0) or torch.any(cur >= dims):
          break
        dScore = step
        if mt[cur[0], cur[1]] >= h:
          break
      cellScore *= dScore
    if p2 < cellScore:
      p2 = cellScore
p2
```
